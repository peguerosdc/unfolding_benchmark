```python
import glob

files = glob.glob("/storage/cpegueros/tau2pipi0_taupair_1623721564_00/sub00/*.root")
print(f"{len(files)} files")
```

```python
import uproot
# check amount of events in each file
total = 0
counts = []
for f in files:
    it = uproot.open(files[0])["tau1x1"]
    count = it.numentries
    print(f"{f}\t{count}")
    counts.append( (f,count) )
    total += count
print(f"TOTAL: {total}")
```

```python
from itertools import islice
# define sizes
total_size = len(counts)
optim = int(.3*total_size)
prep = int(.3*total_size)
data = int(.4*total_size)
print(f"optim={optim}, prep={prep}, data={data}")
# split
sizes = [optim, prep, data]
countsi = iter(counts)
chunks = [list(islice(countsi, None, i)) for i in sizes]
```

```python
# print elements in each chunk
this_total = 0
for size, chunk in zip(sizes, chunks):
    in_chunk = 0
    for (f, count) in chunk:
        in_chunk += count
    print(f"Chunk with {size/total_size} has {in_chunk/total} = {in_chunk}")
    this_total += in_chunk
print(f"Totals match? {this_total == total}")
```
